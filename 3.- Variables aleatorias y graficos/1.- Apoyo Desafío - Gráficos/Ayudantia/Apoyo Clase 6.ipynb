## Fórmula vectorizada en nueva columna

```python
df = pd.DataFrame([
    {"nombre": "Neko", "edad": 17, "especie": "gato", "cant_juguetes": np.nan}, 
    {"nombre": "Fantasma", "edad": 5, "especie": "erizo", "cant_juguetes": 2}, 
    {"nombre": "Miti", "edad": 10, "especie": "gato", "cant_juguetes": np.nan},
    {"nombre": "Bahamut", "edad": 1, "especie": "ratón", "cant_juguetes": 5}, 
    {"nombre": "Maggie", "edad": 7, "especie": "perro", "cant_juguetes": 3}, 
    {"nombre": "Alonso", "edad": 2, "especie": "perro", "cant_juguetes": 6}
])
df
```

```python
# a ** b; a = base; b = exponente
df["edad_cuadrado"] = df["edad"] ** 2
df
```

```python
df["puntaje_giani"] = ((df["edad"] + 10) / 2) ** 2
df
```

## Uso de legend

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

x_axis = np.linspace(.3, 1, 1000)
plt.title("Uso de label y legend")
plt.plot(x_axis, stats.norm.pdf(x_axis, 0.696, np.sqrt(0.024)), color='pink', lw=3, label="Curva rosa")
plt.plot(x_axis, stats.norm.pdf(x_axis + 0.05, 0.696, np.sqrt(0.024)), color='violet', lw=3, label="Curva violeta")
plt.legend();
```
