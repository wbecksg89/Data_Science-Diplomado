```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
columns_float = df.dtypes[df.dtypes=='float64'].index

nombre = df[columns_float].isnull().sum().sort_values().index[0]

df = df.dropna(subset=[nombre])

std = df[nombre].std()
mean_mu = df[nombre].mean()

df['fare_norm'] = [(i - mean_mu)/std for i in df[nombre]]
df = df.sort_values(by='fare_norm')
df.head(10)
```
