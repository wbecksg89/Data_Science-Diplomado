### Recodificar en variables nuevas

```python
next(iter(enumerate(df['Age'])))
```

```python
for i,edad in enumerate(df["Age"][1:10]):
    print(i,edad)
```

#### Forma 1

```python
df["nueva"] = [1 if x>df["Age"].median() else 0 for x in df["Age"]]
df.head()
```

#### Forma 2

```python
mediana = df["Age"].median()
def reco_new(fila):
    if fila > mediana:
        return 1
    else:
        return 0
    
df["nueva3"] =df["Age"].map(reco_new)
df.head()
```

#### Forma 3

```python
df["nueva2"] = 0
mediana = df["Age"].median()
for i,edad in enumerate(df["Age"]):
    if edad > mediana:
        df.at[i,"nueva2"] = 1

df.head(5)
```

#### Forma 4 

```python
df["nueva4"] = df["Age"].apply(lambda x: reco_new(x))
df.head()
```

### Agrupar y contar

```python
var = ["nueva","nueva2","nueva3","nueva4"]

for i in var:
    print(df[i].value_counts(), "\n")
```
