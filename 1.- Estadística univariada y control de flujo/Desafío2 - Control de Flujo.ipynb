### Desafío 3
Genere un loop con el segundo array que cuente las siguientes condiciones:
1. Si el número es divisible por 2 o 3.
2. Si el número es divisible por 2 y 3.
3. Si el número es divisible por 3 pero no por 2.
4. si el número no es divisible por 2 ni 3.


```python
count1=0
count2=0
count3=0
count4=0

for i in arr2:
    if i%2==0 or i%3==0:
        count1+=1
    if i%2==0 and i%3==0:
        count2+=1
    if i%2!=0 and i%3==0:
        count3+=1
    if i%2!=0 and i%3!=0:
        count4+=1
print("1. Si el número es divisible por 2 o 3: ",count1)
print("2. Si el número es divisible por 2 y 3: ",count2)
print("3. Si el número es divisible por 3 pero no por 2:", count3)
print("4. si el número no es divisible por 2 ni 3: ", count4)
```

## Desafío 5
* Utilizando la misma base flights.csv hacer un loop y clasificar los meses con una
cantidad de pasajeros menor a la media.
* Para ello, generen un nuevo objeto que represente la media de passengers.
* Generen una columna en la base de datos que se llame underperforming y asígnele
0.
* Ejecuten un loop que recorra cada observación de passengers, donde si la
observación es menor a la media de passengers se le asigne a underperforming
un 1.
* Para asignar unos en la columna underperforming ya creada, para ello utilice la
función at de pandas.


```python
import pandas as pd
df = pd.read_csv('flights.csv')
df['underperforming'] = 0
media = df['passengers'].mean()
for i,j in enumerate(df["passengers"]):
    if j < media:
        df.at[i,"underperforming"] = 1

df
```

```python
df['menor_a_la_media'] = df.apply(lambda x: 1 if x['passengers'] < df['passengers'].mean() else 0,1)
df
```

## Desafío 6
Hacer un loop que clasifique los meses donde la cantidad de pasajeros se escapa de la
tendencia.
* Para ello, genere dos objetos que guarden la media general y la desviación estándar
general de passengers.
* Genere una nueva columna en la tabla de datos que se llame outlier y asignele 0.
* Ejecuten un loop que recorra cada observación de passengers, donde si la
observación se escapa de la tendencia sea 1, de lo contrario 0.
* Para clasificar los casos que se escapen la tendencia, la observación debe satisfacer
una de las siguientes condiciones:
    * La observación debe ser menor a la media menos la desviación estándar
    * La observación debe ser mayor a la media más la desviación estándar.
    * ¿Cuántas observaciones se pueden clasificar como casos extremos?


```python
df['media_general'] = df['passengers'].mean()
df['std'] = df['passengers'].std()
df['max_tend'] = df['passengers'].mean()+df['passengers'].std()
df['min_tend'] = df['passengers'].mean()-df['passengers'].std()
df['outliers'] = 0
for i,j in enumerate(df["passengers"]):
    if df['passengers'].mean()+df['passengers'].std() < j  or df['passengers'].mean()-df['passengers'].std()>j:
        df.at[i,"outliers"] = 1
df.tail(30)
```
